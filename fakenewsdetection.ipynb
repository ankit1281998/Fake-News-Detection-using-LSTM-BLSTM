```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fake=pd.read_csv("Fake.csv")
fake
```

```python
true=pd.read_csv("True.csv")
true
```

```python
fake["target"]="fake"
true["target"]="true"
```

```python
fake.head()
```

```python
data=pd.concat([fake,true]).reset_index(drop="index")
data
```

```python
from sklearn.utils import shuffle
data=shuffle(data)
data
```

```python
data=data.reset_index(drop="index")
data
```

```python
data.info()
```

```python
data=data.drop(columns=["date","title"])
data
```

```python
cleaned_data=data.copy()
```

```python
data["text"]=data["text"].apply(lambda x:x.lower())
data
```

```python
import string

def punctuation_removal(text):
    all_list = [char for char in text if char not in string.punctuation]
    clean_str = ''.join(all_list)
    return clean_str

data['text'] = data['text'].apply(punctuation_removal)
```

```python
import nltk
from nltk.corpus import stopwords
stop=stopwords.words("english")
```

```python
data["text"]=data["text"].apply(lambda x:" ".join([word for word in x.split() if word not in (stop)]))
data
```

# Model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, y_train, y_test= train_test_split(data["text"],data["target"],test_size=0.2,random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python

tv=TfidfVectorizer(max_features=1500)
X2=tv.fit_transform(X_train).toarray()
X2
```

```python
tv=TfidfVectorizer(max_features=1500)
X3=tv.fit_transform(X_test).toarray()
X3
```

```python
model=DecisionTreeClassifier()
model.fit(X2,y_train)
```

```python
prediction=model.predict(X3)
prediction
```

```python
accuracy_score(y_test,prediction)
```

```python
y_test[:10],prediction[:10]
```

```python
cm=confusion_matrix(y_test,prediction)
cm
```
